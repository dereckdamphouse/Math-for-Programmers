```python
def eulers_method(s0,v0,a,time,steps):
    s = s0
    v = v0
    positions = [s0]
    dt = time/steps
    for _ in range(0,steps):  
        s = add(s, scale(dt,v))
        v = add(v, scale(dt,a))
        positions.append(s)
    return positions
```

```python
e1 = eulers_method((0,0),(1,0),(0,0.2),10,10)
```

```python
def eulers_method_over_y(s0,v0,a,time,steps):
    s = s0
    v = v0
    positions = [s0]
    dt = time/steps
    for _ in range(0,steps):  
        v = add(v, scale(dt,a))
        s = add(s, scale(dt,v))
        positions.append(s)
    return positions
```

```python
e2 = eulers_method_over_y((0,0),(1,0),(0,0.2),10,10)
```

```python
e3 = eulers_method((0,0),(1,0),(0,0.2),10,100)
```

```python
draw2d(Points2D(*e1))
```

```python
draw2d(Points2D(*e2))
```

```python
draw2d(Points2D(*e3))
```

```python
from math import cos, sin, pi
```

```python
def distance_traveled(s0,speed,angle,time,steps):
    angle = angle * pi/180
    s = s0
    v = (speed * cos(angle), speed * sin(angle))
    a = (0, -9.81)
    dt = time/steps
    for _ in range(0,steps):
        s = add(s, scale(dt,v))
        v = add(v, scale(dt,a))
        if s[1] <= 0:
            return round(s[0])
```

```python
distance_traveled((0,1.5),30,20,10,100)
```

```python
s = (0,0)
angle = 10
```

```python
distance_traveled(s,30,angle,10,100)
```
