```python
def addv(*vectors):
    return (sum([v[0] for v in vectors]), sum([v[1] for v in vectors]))
```

```python
def translate(translation, vectors):
    return [addv(translation, v) for v in vectors]
```

```python
def subtract(v1, v2):
    return (v1[0]-v2[0], v1[1]-v2[1])
```

```python
def distance(v1, v2):
    d = subtract(v1, v2)
    return sqrt(d[0]**2 + d[1]**2)
```

```python
def perimeter(vectors):
    distances = [
        distance(
            vectors[i],
            vectors[(i+1)%len(vectors)]
        )
        for i in range(0,len(vectors))
    ]
    return sum(distances)
```

```python
from math import atan2
def to_polar(vector):
    x, y = vector[0], vector[1]
    angle = atan2(y,x)
    return(length(vector), angle)
```

```python
def rotate(angle, vectors):
    polars = [to_polar(v) for v in vectors]
    return [to_cartesian((l, a + angle)) for l,a in polars]
```

```python
draw(
    Polygon(*dino_vectors, color=gray),
    Polygon(*rotate(pi/4, dino_vectors), color=red)
)
```

```python
def regular_polygon(n):
    angle = 2*pi/n
    points = [to_cartesian((1,angle*x)) for x in range (0,n)]
    draw(
        Points(*points),
        Polygon(*points)
    )
```

```python
regular_polygon(7)
```

```python
regular_polygon(100)
```
